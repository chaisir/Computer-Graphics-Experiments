```python
import matplotlib.pyplot as plt
import math

def draw_4points(x,y,xc,yc):
    plt.scatter (x+xc,y+yc)
    plt.scatter (x+xc,-y+yc)
    plt.scatter (-x+xc,-y+yc)
    plt.scatter (-x+xc,y+yc)
    ;
    
    
def draw3(a,b,xc,yc):              #中点画椭圆
    plt.figure(1)
    plt.title('Bresenham_Ellipse')
    plt.axis('equal')
    
    
    d = b*b + a*a*(-b + 0.25);
    x = 0
    y = b
    draw_4points(x,y,xc,yc)
    
    fx =int((a*a) / math.sqrt(a*a + b*b)) 
    while (x != fx):
        if (d < 0):
            d += b*b*(2 * x + 3)
        else:
            y-=1
            d += b*b*(2 * x + 3) + a*a*(-2 * y + 2)
        x+=1
        draw_4points(x,y,xc,yc)
        

    # 中点改为(x+0.5,y-1)所以d = 
    d = b*b*(x + 0.5)*(x + 0.5) + a*a*(y - 1)*(y - 1) - a*a*b*b
    while (y > 0):
        if (d < 0):
            x+=1
            d += b*b*(2 * x + 2) + a*a*(-2 * y + 3)

        else:
            d += a*a*(-2 * y + 3)
        y-=1
        draw_4points(x,y,xc,yc)

    
    ;
    
    

a = int(input("长轴半径："))
b = int(input("短轴半径："))
xc = int(input("椭圆圆心横坐标："))
yc = int(input("椭圆圆心纵坐标："))
draw3(a,b,xc,yc)
```
