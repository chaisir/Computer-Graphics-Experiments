```python
import matplotlib.pyplot as plt


def draw1(x1, y1, x2, y2):  # 数值微分法
    plt.figure(1)
    plt.title('DDA_line')
    plt.axis('equal')
    dx = x2 - x1
    dy = y2 - y1
    if (abs(dx) - abs(dy) > 0):  # 比较两参数的绝对值哪一个大，哪一个就作为步长参数（n），
        n = abs(dx)  # 此参数将作为沿直线所画出点的数目
    else:
        n = abs(dy)
    xinc = dx / n
    yinc = dy / n
    x = x1
    y = y1
    plt.scatter(x,y)
    for k in range(n):
        plt.scatter(int(x + 0.5), int(y + 0.5))
        x += xinc
        y += yinc
#     plt.show()

def draw2(x0, y0, x1, y1):  # 中点画线法
    plt.figure(2)
    plt.title('Midpoint_line')
    plt.axis('equal')

    if (x1 < x0):
        t = x1
        x1 = x0
        x0 = t
        q = y1
        y1 = y0
        y0 = q

    a = y0 - y1
    b = x1 - x0

    if (b==0):
        for y in range(y0,y1):
            plt.scatter(x0, y)
    else:
        k=-a/b
        x = x0
        y = y0
        plt.scatter(x, y)

        if (k >= 0 and k <= 1):
            d = 2 * a + b
            while(x<x1):
                if (d <= 0):
                    x+=1
                    y += 1
                    d += 2 * (a + b)
                else:
                    x+=1
                    d += 2 * a
                plt.scatter(x, y)

        elif (k > 1):
            d = a + 2 * b
            for y in range(y0,y1):
                if (d < 0):
                    d += 2 * b
                else:
                    x += 1
                    d += 2 * (a + b)
                plt.scatter(x, y)

        elif (k >= -1 and k < 0):
            d = 2 * a - b
            for x in range(x0,x1):
                if (d < 0):
                    d += 2 * a
                else:
                    y -= 1
                    d += 2 * (a - b)
                plt.scatter(x, y)

        elif (k < -1):
            d = a - 2 * b
            for y in range(y0,y1,-1):
                if (d < 0):
                    x += 1
                    d += 2 * (a - b)
                else:
                    d -= 2 * b
                plt.scatter(x, y)

#     plt.show()



def draw3(x0, y0, x1, y1):  # Bresenham算法
    plt.figure(3)
    plt.title('Bresenham_line')
    plt.axis('equal')
    
    dx = x1 - x0
    dy = y1 - y0
    if(dx==0):
        if(y0>y1):
            t = x1
            x1 = x0
            x0 = t
            q = y1
            y1 = y0
            y0 = q

        for y in range(y0,y1):
            plt.scatter(x0, y)
    else:
        k=dy/dx
        if(k>=0 and k<=1):
            if(x0>x1):
                t = x1
                x1 = x0
                x0 = t
                q = y1
                y1 = y0
                y0 = q
            d=0.5-k
            x=x0
            y=y0
            for x in range(x0,x1):
                plt.scatter(x,y)
                if (d<0):
                    y+=1
                    d+=1-k
                else:
                    d-=k
        if(k>1):
            if(y0>y1):
                t = x1
                x1 = x0
                x0 = t
                q = y1
                y1 = y0
                y0 = q
            d=1-0.5*k
            x = x0
            y = y0
            for y in range(y0, y1):
                plt.scatter(x, y)
                if (d >= 0):
                    x += 1
                    d += 1 - k
                else:
                    d += 1

        if (k >= -1 and k<=0):
            if (x0 > x1):
                t = x1
                x1 = x0
                x0 = t
                q = y1
                y1 = y0
                y0 = q
            d = -0.5-k
            x = x0
            y = y0
            for x in range(x0, x1):
                plt.scatter(x, y)
                if (d > 0):
                    y -= 1
                    d -= 1 + k
                else:
                    d -= k

        if (k < -1):
            if (y0 < y1):
                t = x1
                x1 = x0
                x0 = t
                q = y1
                y1 = y0
                y0 = q
            d = -1 - 0.5 * k
            x = x0
            y = y0
            for y in range(y0, y1,-1):
                plt.scatter(x, y)
                if (d < 0):
                    x += 1
                    d -= 1 + k
                else:
                    d -= 1

#     plt.show()


x1 = int(input("起点横坐标:"))
y1 = int(input("起点纵坐标:"))
x2 = int(input("终点横坐标:"))
y2 = int(input("终点纵坐标:"))
draw1(x1, y1, x2, y2)
draw2(x1, y1, x2, y2)
draw3(x1, y1, x2, y2)
```
